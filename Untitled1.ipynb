```python
#Filtrado de un electrocardiograma

#Se usó como guia https://programmerclick.com/article/6115655816/

#import numpy as np
#import pip
#pip.main(["install","matplotlib"])
#pip.main(["install","scipy"])

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import scipy.fftpack as fourier
from scipy.signal import butter, sosfilt, sosfilt_zi, sosfiltfilt, lfilter, lfilter_zi, filtfilt, sosfreqz, resample
```

```python
#Data extraída de https://github.com/marianpetruk/ECG_analysis

import os

data_path = os.path.join("data", "Lviv-Biometric-Data-Set")
data_dirs = os.listdir(data_path)
print(len(data_dirs), "users = ", data_dirs)

USERID = 1
user_path = os.path.join(data_path, "user" + str(USERID))

user_ecg = os.listdir(user_path)
print("\nuser" + str(USERID), "has", len(user_ecg), "ecg samples =", user_ecg)
```

```python
SAMPLEID = 0
data = np.load(os.path.join(user_path, user_ecg[SAMPLEID]))
print(type(data))
print(data)
print(data.shape)
```

```python
#La señal de un electrocardiograma se puede considerar como una función dependiente del tiempo.
#Cuando se tienen este tipo de funciónes puede realizarse una Transformada de Fourier con el fin
#de visualizar cuale es el dominio en términos de frecuencia.

#Entonces, primero veamos la data y su respectiva FFT. 
```

```python
#Plot de la data

plt.figure(figsize=(20, 10))
plt.ylabel("Amplitude (dB)")
fs = 200

times = np.arange(data.shape[0], dtype='float') / fs
plt.plot(times, data)
plt.xlabel("Time (s)")
plt.xlim(0,15)
plt.title("ECG sample raw data")
plt.show()
```

```python
#Transformada de Fourier

fun_fft = fourier.fft(data)                           # Calculo de la FFT
m_fft = abs(fun_fft)                                  # Calculo la la Magnitud de la FFT
vf = fs*np.arange(0, len(data))/len(data)             # Definición del Vector de Frecuencias


plt.figure(figsize=(20, 5))
plt.ylabel("Power")
fs = 200

times = np.arange(data.shape[0], dtype='float') / fs
plt.plot(vf, m_fft)
plt.xlabel("Frecuency (Hz)")
#plt.xlim(-4,200)
#plt.ylim(-4,300000)
plt.title("FFT")
plt.show()
```

```python
#En el plot de la FFT, se ve que hay un pico intenso alrededor de los 36/37Hz, el cual es debido a un efecto de
#superposicion de ondas de la señal original, que se llama interferencia de frecuencia de potencia. 

#Esta frecuencia es la misma con la que varia la corriente alterna de la red eléctrica. 
#La corriente alterna fluye a través del cable de red e irradia energía hacia el espacio. 
#Esta energía radiada ingresa al amplificador de señal a través de la capacidad de acoplamiento del cuerpo humano,
#el cable del electrodo y el cable de red, y forma un "amplificador de señal".  
#Por otro lado, puede verse un pico en 72/73Hz, el cual es solo una duplicación de frecuencia de 36/37Hz, 
#que también es parte de la llamada interferencia de frecuencia de potencia. 


#También puede verse que hay energía alrededor de los 0Hz, y estos componentes de baja frecuencia corresponden 
#a la línea de base que cambia lentamente en la señal original. 

#Las estructuras antes mencionaas no son parte del diagnóstico médico, por lo tanto, 
#es importante que sean filtrados. 
```

```python
#Entonces, se necesitan dos filtros, uno de los cuales es un filtro de paso de banda de 3-60 Hz 
#(estos son puestos a ojo en base al articulo que se uso como guia),
#que retiene los componentes de frecuencia de 3-60 Hz en la señal, es decir,
#se elimina la parte de baja frecuencia por debajo de 3 Hz y la parte de alta frecuencia por encima de 60 Hz. 

#El otro es un filtro que no deje pasar una frecuencia especifica, en este caso 35-40Hz, 
#que elimina el componente de 36-37Hz de la señal.
```

```python
#Filtro correspondiente al paso de banda 3-60 Hz
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    sos = butter(order, [low, high], analog=False, btype="band", output="sos")
    return sos

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    sos = butter_bandpass(lowcut, highcut, fs, order=order)
    y = sosfilt(sos,
                data)  
    return y

#Filtro correspondiente a la componente componente de 36-37Hz
def butter_bandstop(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    sos = butter(order, [low, high], analog=False, btype="bandstop", output="sos")
    return sos

def butter_bandstop_filter(data, lowcut, highcut, fs, order=5):
    sos = butter_bandstop(lowcut, highcut, fs, order=order)
    z = sosfilt(sos,
                data)  
    return z


x = data
fs = 400 
lowcut = 3
highcut = 60


# Plot de la respuesta en frecuencia del primer filtro
plt.figure(1, figsize=(20, 8))
plt.clf()
for order in [2, 3, 4]:
    sos = butter_bandpass(lowcut, highcut, fs, order=order)
    w, h = sosfreqz(sos, worN=2000)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         "--", label="sqrt(0.5)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain")
plt.title("Frequency response for a few different orders")
plt.grid(True)
plt.legend(loc="best")

#Las siguientes figuras muestran la respuesta de frecuencia de ambos filtros. 
#El eje horizontal es la frecuencia, y el eje vertical es la ganancia del filtro a la señal de frecuencia.
#Cuando la ganancia es 1.0, significa que la señal de frecuencia no está obstruida a través del filtro. 
#Cuando la ganancia es menor a 1.0, significa que la señal de frecuencia pasa a través del filtro. 
#Estar atenuado Se puede ver que cuanto mayor sea el orden de orden del filtro, 
#mejor será el rendimiento del filtro (la curva de respuesta de frecuencia es empinada).


#En el código, se generaron 3 filtros de orden 2,3,4 para cada tipo de filtro. 
#En la figura, también se dibuja una línea punteada de sqrt (0.5), que corresponde 
#a la llamada ganancia de -3dB. Es donde, la potencia de la señal correspondiente a la frecuencia 
#se reduce a la mitad de la potencia más alta.
```

```python
# Plot de la respuesta en frecuencia del segundo filtro
plt.figure(1, figsize=(20, 8))
plt.clf()
for order in [2, 3, 4]:
    sos = butter_bandstop(35, 40, fs, order=order)
    w, h = sosfreqz(sos, worN=2000)
    plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
         "--", label="sqrt(0.5)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain")
plt.xlim(0,100)
plt.title("Frequency response for a few different orders")
plt.grid(True)
plt.legend(loc="best")
```

```python
#Ahora se aplican los filtros por separado a la señal original

x = data[:50000]
plt.figure(figsize=(20, 8))

plt.ylabel("Amplitude (dB)")

times = np.arange(x.shape[0], dtype='float') / fs
plt.xlabel("Time (s)")

plt.plot(times, x, "g", linewidth=0.5, label="Señal original")
plt.legend(loc="lower left")
plt.twinx()
plt.plot(times, butter_bandpass_filter(x, lowcut, highcut, fs, order=4),
         label="Señal filtrada (%g Hz) con el paso de banda entre 3 y 60 Hz" % fs)
plt.plot(times, butter_bandstop_filter(x, lowcut, highcut, fs, order=4),
         label="Señal filtrada (%g Hz) de la frecuencia 36/37 Hz" % fs)
plt.grid(True)
plt.axis("tight")
plt.legend(loc="upper right")

plt.show()
```

```python
#Aplico ambos filtrados a la señal

x = data[:500]
# Filtrado de paso de banda
a = butter_bandpass_filter(x,3 ,60,fs,order=4)

# Filtrado de la frecuencia 36/37 Hz
b= butter_bandstop_filter(a,35,40,fs, order=4)


plt.figure(figsize=(20, 8))
plt.ylabel("Amplitude (dB)")
times = np.arange(b.shape[0], dtype='float') / fs
plt.xlabel("Time (s)")
plt.plot(times, b, "g", linewidth=0.5, label="Señal con ambos filtros")
plt.legend(loc="lower left")
plt.twinx()
```

```python
#Ahora miro como quedo la FFT de la señal filtrada

fun2_fft = fourier.fft(b)                           # Calculo de la FFT
m_fft2 = abs(fun2_fft)                              # Calculo la la Magnitud de la FFT
vf2 = fs*np.arange(0, len(b))/len(b)                # Definición del Vector de Frecuencias


plt.figure(figsize=(20, 5))
plt.ylabel("Power")
fs = 200

times = np.arange(b.shape[0], dtype='float') / fs
plt.plot(vf2, m_fft2)
plt.xlabel("Frecuency (Hz)")
plt.xlim(-4,80)
#plt.ylim(-4,300000)
plt.title("FFT")
plt.show()
```
